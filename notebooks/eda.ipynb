Correlation Analysis

```python
num_cols = df.select_dtypes(include='number').columns
num_cols
```

```python
corr = df[num_cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr,annot=False,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Identifying Missing Values

```python
df.isnull().sum()
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)
```

Outlier Detection

```python
df[num_cols].boxplot(figsize=(10,6))
plt.title("Boxplot for all the numerical values")
plt.show()
```

```python
for coli in num_cols:
    skewn = df[coli].skew()
    print(skewn)
```

Insights Based On Finidings

```python
print("===SKEWNESS===")
print("\n The Amount, Value, and FraudResult features exhibit Extreme Positive Skewness \nwith the skewness of 51.09 for Amount , 51.29 for Value and 22.19 for FraudResult ")
print("This skewness indicates that the mean of the data is significantly higher than the median due to a few very high-value transactions pulling the average upward")
print("\n===OUTLIERS===")
print("\nThe box plot confirms significant outliers in the Amount and Value columns.")
print("\n===CORRELATION ANALYSIS")
print("\nThere is a perfect positive correlation between Amount and Value. This suggests that Amount and Value are essentially the same data or very closely derived from one another")
print("There is a weak positive correlation between transaction size and fraud.\n While higher amounts might be slightly more associated with fraud, the relationship isn't strong enough to use as a sole predictor. \nFraud detection will likely require a combination of multiple features.")
print("There is a negative correlation between PricingStrategy with Amount, Value, and FraudResult")
print("as the PricingStrategy code increases, the Amount or likelihood of a FraudResult tends to decrease slightly or vice versa.")

```
